```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Model, layers, optimizers, datasets
```

```python
(images_train, labels_train), (images_test, labels_test) = datasets.cifar10.load_data()
```

```python
print(images_train.shape)
print(images_test.shape)
```

```python
images_train = images_train.astype('float32')/255
images_test = images_test.astype('float32')/255
```

```python
input_shape = (32, 32, 3)
num_classes = 10
```

```python
model = keras.models.Sequential([
  layers.Conv2D(32, (3,3), padding='same', activation='relu', input_shape=input_shape),
  layers.MaxPool2D(strides=(2,2)),
  layers.BatchNormalization(),

  layers.Conv2D(64, (3,3), padding='same', activation='relu'),
  layers.MaxPool2D(strides=(2,2)),
  layers.BatchNormalization(),

  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.BatchNormalization(),
  layers.Dense(num_classes, activation='softmax')
])
```

```python
model.complie(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model_train = model.fit(images_train, labels_train, batch_size=128, epochs=3, validation_split=0.2)
```

```python
test_loss, test_acc = model.evaluate(images_test, labels_test, verbose=2)
print('test accuracy:', test_acc)
```
