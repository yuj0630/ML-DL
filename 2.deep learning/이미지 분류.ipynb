```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import datasets
```

```python
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()
```

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

x_train = x_train/255
x_test = x_test/255
```

```python
flat_x_train = x_train.reshape(x_train.shape[0], -1)
flat_x_test = x_test.reshape(x_test.shape[0], - 1)

print(flat_x_train.shape)
```

```python
y_train = y_train.reshape(y_train.shape[0])
y_test = y_test.reshape(y_test.shape[0])
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
#KNN
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(flat_x_train, y_train)
```

```python
y_pred_knn = knn.predict(flat_x_test)

from sklearn.metrics import accuracy_score
```

```python
test_accuracy = accuracy_score(y_test, y_pred_knn)
```

```python
test_accuracy
```

```python
random_forest = RandomForestClassifier(criterion='entropy', max_depth=70, n_estimators=100)
random_forest.fit(flat_x_train, y_train)
```

```python
y_pred_forest = random_forest.predict(flat_x_test)
test_accuracy = accuracy_score(y_test,  y_pred_forest)
print(test_accuracy)
```

```python
tree = DecisionTreeClassifier(max_depth=100, criterion='entropy')
tree.fit(flat_x_train, y_train)
```

```python
y_pred_tree = tree.predict(flat_x_test)
test_accuracy = accuracy_score(y_test, y_pred_tree)
print(test_accuracy)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=500)
pca.fit(flat_x_train)
```

```python
x_train_pca = pca.transform(flat_x_train)
x_test_pca = pca.transform(flat_x_test)
print(x_train_pca.shape)
```

```python
x_test_pca.shape
```

```python
# KNN
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_pca, y_train)
```

```python
y_pred = knn.predict(x_test_pca)
knn_pca_accuracy = accuracy_score(y_test, y_pred)
print(knn_pca_accuracy)
```

```python
#Random Forest
```

```python
random_forest = RandomForestClassifier(criterion='entropy', max_depth=70, n_estimators=100)
random_forest.fit(x_train_pca, y_train)
```

```python
y_pred = random_forest.predict(x_test_pca)
random_forest_pca_accuracy = accuracy_score(y_test, y_pred)
random_forest_pca_accuracy
```

```python
#Decision Tree
```

```python
tree = DecisionTreeClassifier(max_depth=100, criterion='entropy')
tree.fit(x_train_pca, y_train)
```

```python
y_pred = tree.predict(x_test_pca)
tree_pca_accuracy = accuracy_score(y_test, y_pred)
tree_pca_accuracy
```
