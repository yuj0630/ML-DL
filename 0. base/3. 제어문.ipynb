## for
list, tuple, str과 같이 인덱싱을 할 수 있는 자료형을 이용해서 반복문을 실행

````python
lst = [1, 2, 3]

for i in lst:
  print(i)
````

````python
[ i for i in lst ]
````

````python
tup = ('a', 'b', 'c')

for i in tup:
  print(i)
````

````python
[ i for i in ('a', 'b', 'c') ]
````

````python
string = "가나다"

for i in string:
  print(i)
````

````python
[ i for i in "가나다" ]
````

## for - continue, break, pass
- pass:   
실행할 코드가 없는 것으로 다음 행동을 계속해서 진행합니다.
- break:    
반복문(loop)을 멈추고 loop 밖으로 나가도록 합니다.
- continue:   
바로 다음 순번의 loop를 수행합니다.

````python
lst = [1,2,3,4,5,6,7]
lst
````

````python
for i in lst:
  if i % 3 == 0: 
    print(f'start')
    pass
    print(f'3의 배수: {i}\n')
  else:
    print(f'i: {i}\n')
````

````python
for i in lst:
  if i % 3 == 0: 
    print(f'start')
    break
    print(f'3의 배수: {i}\n')
  else:
    print(f'i: {i}\n')
````

````python
for i in lst:
  if i % 3 == 0: 
    print(f'start')
    continue
    print(f'3의 배수: {i}\n')
  else:
    print(f'i: {i}\n')
````

## for - else
for문 실행이 완료되면 else가 실행된다.

````python
for i in lst:
  if i % 2 == 0: # 짝수만...
    pass
    print(f'짝수: {i}')
  else:
    print(f'홀수: {i}')
else:
  print(f'for문 완료')
````

````python
for i in lst:
  if i % 2 == 0: 
    continue
    print(f'짝수: {i}')
  else:
    print(f'홀수: {i}')
else:
  print(f'for문 완료')
````

for문에서 중간에 멈추면(break) else는 실행되지 않는다.

````python
for i in lst:
  if i % 2 == 0: 
    break
    print(f'짝수: {i}')
  else:
    print(f'홀수: {i}')
else:
  print(f'for문 완료')
````

# 예외처리

## try/except
```python
try:
  # 로직, 기능 
except <오류 종류>:
  # try문에서 Exception(오류)가 발생할 때, 실행됨
```

````python
try:
  x = int('11')
  print(x)
except ValueError as e:
  print(f'오류 메세지: {e}')
````

````python
try:
  x = int('abc')
  print(x)
except ValueError as e:
  print(f'오류 메세지: {e}')
````

````python
try:
  int('abc')
except Exception as e:
  print(f'오류명: {type(e)} / 오류 메세지: {e}')
````

````python
try: 
  x = 11 + int('abc')
  print(x)
except (ValueError, TypeError) as e:
  print(f'오류명: {type(e)} / 오류 메세지: {e}')
````

````python
try:
  x = [1, 2, 3]
  x.append(4)
  print(x)
  x + 'aa'
except ValueError as e:
  print(f'오류명: {type(e)} / 오류 메세지: {e}')
except TypeError as e:
  print(f'오류명: {type(e)} / 오류 메세지: {e}')
````

````python
try:
  x = [1, 2, 3]
  x.appe(4)
except ValueError as e:
  print(f'오류명: {type(e)} / 오류 메세지: {e}')
except TypeError as e:
  print(f'오류명: {type(e)} / 오류 메세지: {e}')
except Exception as e:
  print(f'오류명: {type(e)} / 오류 메세지: {e}')
````

## else/finally
```python
try:
  # 로직, 기능 
except <오류종류>:
  # try문에서 오류가 발생할때, 실행 
else:
  # try문이 성공할 때 발생 
finally:
  # try문의 결과와 상관없이 실행
```

````python
try:
  x = 11 + int('22')
  print(x)
except:
  print('except ~~')
````

````python
try:
  x = 11 + int('22')
  print(x)
except:
  print('except ~~')
finally:
  print('finally ~~')
````

````python
try:
  x = 11 + int('22')
  print(x)
except:
  print('except ~~')
else:
  print('else ~~')
finally:
  print('finally ~~')
````

````python
try:
  x = 11 + int('abc')
  print(x)
except:
  print('except ~~')
else:
  print('else ~~')
finally:
  print('finally ~~')
````

## assert
- 조건이 False일 때, 발생하는 에러 
- 디버깅 모드에서만 작동하지만, 파이썬은 기본 디버깅 모드임 

````python
x = 3

assert x % 3 == 0, '3의 배수가 아닙니다.'

print('3의 배수입니다.')
````

````python
x = 5

assert x % 3 == 0, '3의 배수가 아닙니다.'

print('3의 배수입니다.')
````

## raise error

````python
try:
  x = 5
  if x % 3 != 0:
    raise Exception('3의 배수가 아닙니다.')
  print('3의 배수 입니다.')
except Exception as e:
  print(f'에러 메세지: {e}')
````

## Exception 만들어서 적용하기 

````python
class NotThreeMultipleError(Exception):
  def __init__(self, error_msg):
    super().__init__(error_msg)
````

````python
try:
  x = 5
  if x % 3 != 0:
    raise NotThreeMultipleError('3의 배수가 아닙니다.')
  print('3의 배수 입니다.')
except Exception as e:
  print(f'오류명: {type(e)} / 오류 메세지: {e}')
````
